**👉 ✅ HIERARCHICAL CHUNKING**

````python
from langchain_text_splitters import RecursiveCharacterTextSplitter
from langchain_core.documents import Document
import uuid

# -----------------------------
# 1️⃣ Parent splitter (LARGE)
# -----------------------------
parent_splitter = RecursiveCharacterTextSplitter(
    chunk_size=600,      # large context
    chunk_overlap=20
)

# -----------------------------
# 2️⃣ Child splitter (SMALL)
# -----------------------------
child_splitter = RecursiveCharacterTextSplitter(
    chunk_size=300,      # precise retrieval
    chunk_overlap=10
)

# -----------------------------
# 3️⃣ Create Parent Chunks
# -----------------------------
parent_docs = parent_splitter.split_documents(documents)

print(f"Total Parent Chunks: {len(parent_docs)}")

# Store parent chunks
parent_store = {}

child_docs = []

# -----------------------------
# 4️⃣ Create Child Chunks
# -----------------------------
for parent in parent_docs:

    parent_id = str(uuid.uuid4())
    parent_store[parent_id] = parent
    #This above line you need to remember---refer slide here

    # Split parent into children
    children = child_splitter.split_documents([parent])

    for child in children:
        child.metadata["parent_id"] = parent_id
        child_docs.append(child)

print(f"Total Child Chunks: {len(child_docs)}")
````

```markdown
This cell implements a hierarchical chunking strategy for documents, which is a common technique in Retrieval-Augmented Generation (RAG) to manage context effectively.

-   **`from langchain_text_splitters import RecursiveCharacterTextSplitter`**:
    -   Imports a text splitter that recursively breaks down text into smaller chunks based on a list of characters. This is suitable for maintaining semantic units.
-   **`from langchain_core.documents import Document`**:
    -   Imports the `Document` class, a core LangChain component for representing text content and its metadata.
-   **`import uuid`**:
    -   Imports the `uuid` module, which is used to generate unique identifiers for the parent chunks.

### **1️⃣ Parent splitter (LARGE)**
-   **`parent_splitter = RecursiveCharacterTextSplitter(...)`**:
    -   An instance of `RecursiveCharacterTextSplitter` is created for generating larger, more contextual chunks (parent chunks).
    -   `chunk_size=600`: Each parent chunk will aim for a maximum of 600 characters.
    -   `chunk_overlap=20`: There will be a small overlap of 20 characters between consecutive parent chunks, helping to preserve context across chunk boundaries.

### **2️⃣ Child splitter (SMALL)**
-   **`child_splitter = RecursiveCharacterTextSplitter(...)`**:
    -   Another instance of `RecursiveCharacterTextSplitter` is created for generating smaller, more precise chunks (child chunks).
    -   `chunk_size=300`: Each child chunk will aim for a maximum of 300 characters.
    -   `chunk_overlap=10`: A smaller overlap of 10 characters is used for child chunks.
    -   The rationale is to retrieve small, highly relevant child chunks quickly and then use their associated larger parent chunks for full context when generating responses.

### **3️⃣ Create Parent Chunks**
-   **`parent_docs = parent_splitter.split_documents(documents)`**:
    -   The `documents` (loaded from the PDF in a previous step) are split into larger `parent_docs` using the `parent_splitter`.
-   **`print(f"Total Parent Chunks: {len(parent_docs)}")`**:
    -   Prints the total number of parent chunks created.
-   **`parent_store = {}`**:
    -   An empty dictionary `parent_store` is initialized to store parent documents, mapping their unique IDs to the `Document` objects themselves. This allows for easy retrieval of the full parent context later.
-   **`child_docs = []`**:
    -   An empty list `child_docs` is initialized to store all the smaller child chunks.

### **4️⃣ Create Child Chunks**
-   **`for parent in parent_docs:`**:
    -   The code iterates through each `parent` document created earlier.
    -   **`parent_id = str(uuid.uuid4())`**: A unique `UUID` (Universally Unique Identifier) is generated for each parent document.
    -   **`parent_store[parent_id] = parent`**: The parent document is stored in `parent_store` with its `parent_id` as the key.
    -   **`children = child_splitter.split_documents([parent])`**: Each `parent` document is further split into smaller `children` chunks using the `child_splitter`. Note that `[parent]` is used to pass a single document as a list to `split_documents`.
    -   **`for child in children:`**:
        -   **`child.metadata["parent_id"] = parent_id`**: For each child chunk, its metadata is updated to include the `parent_id` of the larger chunk it originated from. This is the crucial link for hierarchical retrieval.
        -   **`child_docs.append(child)`**: The child chunk is added to the `child_docs` list.
-   **`print(f"Total Child Chunks: {len(child_docs)}")`**:
    -   Prints the total number of child chunks created.

This hierarchical chunking approach allows for efficient retrieval (by searching small, precise child chunks) while still providing a comprehensive context for the Language Model (by retrieving the larger parent chunks associated with the relevant children).
```

````python
def retrieve_parent_context(query):
    """
    Retrieve child chunks → map to parent → return full context
    """
    child_results = retriever.invoke(query)

    parent_ids = set()
    parent_contexts = []

    for child in child_results:
        parent_id = child.metadata["parent_id"]
#two child may have same Parent id
        if parent_id not in parent_ids:
            parent_ids.add(parent_id)

            # ✅ Get full parent document
            parent_doc = parent_store[parent_id]
            parent_contexts.append(parent_doc.page_content)

    return "\n\n".join(parent_contexts)
````

```markdown
This cell defines the `retrieve_parent_context` function, which is a core component of the hierarchical RAG pipeline. This function takes a query, uses the retriever to find relevant child chunks, and then maps those child chunks back to their larger parent contexts.

-   **`def retrieve_parent_context(query):`**:
    -   Defines a function named `retrieve_parent_context` that accepts one argument: `query` (the user's question or input).

    -   It returns a list of the top `k` (e.g., 3, as configured earlier) most semantically similar child chunks based on the query.

-   **`parent_ids = set()`**:
    -   An empty set named `parent_ids` is initialized. A set is used here to store unique parent IDs, ensuring that each parent document is considered only once, even if multiple retrieved child chunks originated from the same parent.

-   **`parent_contexts = []`**:
    -   An empty list named `parent_contexts` is initialized. This list will eventually hold the `page_content` (the actual text) of the unique parent documents that are identified.

-   **`for child in child_results:`**:
    -   The code iterates through each `child` document that was returned by the `retriever`.

    -   **`parent_id = child.metadata["parent_id"]`**:

    -   **`if parent_id not in parent_ids:`**:
        -   This condition checks if the current `parent_id` has already been added to the `parent_ids` set. This prevents duplicate retrieval and processing of the same parent document.

        -   **`parent_ids.add(parent_id)`**:
            -   If the `parent_id` is new, it's added to the `parent_ids` set.

        -   **`parent_doc = parent_store[parent_id]`**:
            -   The full `parent_doc` object is retrieved from the `parent_store` dictionary (which was populated during the hierarchical chunking process) using the `parent_id` as the key.

            -   The actual text content (`page_content`) of the retrieved parent document is appended to the `parent_contexts` list.
```

````python
# ✅ RAG Pipeline (Hierarchical ✅)
rag_chain = (
    {
        "context": retrieve_parent_context,   # ✅ changed here
        "question": RunnablePassthrough()
    }
    | prompt
    | llm
    | StrOutputParser()
)
````

```markdown
This cell defines the core Retrieval-Augmented Generation (RAG) pipeline as a LangChain Expression Language (LCEL) chain. This chain orchestrates the flow from a user's question to a generated answer, integrating retrieval and language model generation.

-   **`rag_chain = (...)`**: This initiates the definition of the RAG chain.

### **Input Processing (First Block)**
-   **`{ "context": retrieve_parent_context, "question": RunnablePassthrough() }`**:
    -   This is a dictionary that defines how the initial input (the user's query) is processed and transformed into the inputs required by the `prompt` template.
    -   **`"context": retrieve_parent_context`**: This is the crucial part of the hierarchical RAG. When the `rag_chain` is invoked with a question, the `retrieve_parent_context` function (defined in the previous cell) is called with the question as its input. This function handles:
        1.  Retrieving relevant `child_docs` using the `retriever`.
        2.  Mapping those `child_docs` back to their larger `parent_docs`.
        3.  Aggregating the `page_content` of these parent documents into a single string.
        The output of this function becomes the value for the `context` key in the dictionary passed to the next step.
    -   **`"question": RunnablePassthrough()`**: This simply takes the original input (the user's question) and passes it through directly as the value for the `question` key. This means the original question is preserved and forwarded.

### **Chain Components (Piped Operations)**
-   **`| prompt`**:
    -   The output from the input processing dictionary (containing `context` and `question` keys) is piped to the `prompt` template. The `prompt` (defined two cells above) then formats these inputs into a structured prompt string for the LLM.

-   **`| llm`**:
    -   The formatted prompt string is then piped to the `llm` (the `ChatOpenAI` model). The LLM processes this prompt and generates a response.

-   **`| StrOutputParser()`**:
    -   Finally, the raw output from the `llm` (which might be a complex object) is piped to `StrOutputParser()`. This parser extracts the actual text content from the LLM's response and converts it into a simple string, making it easy to display or use downstream.

In summary, this `rag_chain` effectively automates the entire process:
1.  Takes a user question.
2.  Finds relevant context hierarchically (small chunks for search, large chunks for content).
3.  Constructs a prompt using this context and the question.
4.  Sends the prompt to the LLM.
5.  Parses the LLM's response into a readable string.
```

```markdown
3.  **`parent_ids = set()`**: An empty set is initialized to store the unique `parent_id`s corresponding to the retrieved child chunks. Using a set ensures that each parent ID is stored only once, even if multiple child chunks belong to the same parent.
4.  **Loop through `child_results`**: The code iterates through each of the retrieved child chunks.
    -   It prints the content of each child chunk to show what was retrieved directly by the `retriever`.
    -   For each child chunk, it extracts its associated `parent_id` from its metadata and adds it to the `parent_ids` set. This step is crucial for the hierarchical retrieval, as it identifies which larger parent documents contain these relevant smaller child chunks.

In essence, this code snippet shows the initial step of finding small, relevant pieces of information and then identifying the larger contexts (parent documents) these pieces came from, preparing for a more comprehensive answer generation.
```

````python
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.output_parsers import StrOutputParser
from langchain_core.runnables import RunnablePassthrough


# ✅ Modern function
def generate_further_insights(
    initial_answer: str,
    prompt_instruction: str,
    llm_model
) -> str:
    """
    Takes an initial answer from RAG and generates further insights
    using a structured prompt pipeline (LCEL).
    """

    print(f"\n--- Performing Secondary Generation: {prompt_instruction} ---")

    # ✅ 1. Prompt Template (Replaces manual f-string)
    prompt = ChatPromptTemplate.from_template("""
Based on the following information:
{initial_answer}

{instruction}
""")

    # ✅ 2. Output Parser
    parser = StrOutputParser()

    # ✅ 3. LCEL Chain
    chain = (
        {
            "initial_answer": RunnablePassthrough(),
            "instruction": lambda x: prompt_instruction
        }
        | prompt
        | llm_model
        | parser
    )

    # ✅ 4. Invoke chain
    response = chain.invoke(initial_answer)

    return response
````
