```python
from langchain_community.document_loaders import TextLoader
from langchain_text_splitters import CharacterTextSplitter

from langchain_openai import OpenAIEmbeddings, ChatOpenAI
from langchain_chroma import Chroma

from langchain_core.runnables import RunnablePassthrough
from langchain_core.prompts import ChatPromptTemplate

# ---------------------------
# 1. Load data
# ---------------------------
loader = TextLoader("/content/ChromaDB/family.txt")
documents = loader.load()

# ---------------------------
# 2. Split
# ---------------------------
text_splitter = CharacterTextSplitter(
    chunk_size=500,
    chunk_overlap=20
)
texts = text_splitter.split_documents(documents)
#✅ Fixed-size / Character-based Chunking
# ---------------------------
# 3. Embeddings
# ---------------------------
embeddings = OpenAIEmbeddings(model="text-embedding-3-small")

# ---------------------------
# 4. Vector DB
# ---------------------------
db = Chroma.from_documents(texts, embeddings)

retriever = db.as_retriever(search_kwargs={"k": 2})

# ---------------------------
# 5. LLM
# ---------------------------
llm = ChatOpenAI(model="gpt-4o-mini", temperature=0)

# ---------------------------
# 6. Prompt
# ---------------------------
prompt = ChatPromptTemplate.from_template("""
Answer the question based only on the context.

Context:
{context}

Question:
{question}
""")

# ---------------------------
# 7. RAG Chain (NEW ✅)
# ---------------------------
rag_chain = (
    {"context": retriever, "question": RunnablePassthrough()}
    | prompt
    | llm
)

# ---------------------------
# 8. Queries
# ---------------------------
queries = [
    "Suhas is having how many years of experience?",
    "What is qualification of Suhas?",
    "Whether Suhas is Post Graduated?"
]

for query in queries:
    response = rag_chain.invoke(query)

    print("\n==========================")
    print("Question:", query)
    print("Answer:", response.content)
```

✅ Strategy Type:
👉 Fixed-size chunking (Character-based splitting)
✅ How it works:

Splits text into chunks of 500 characters
Keeps 20 characters overlap between chunks

Chunk 1 → 0–500 characters
Chunk 2 → 480–980 characters  
Chunk 3 → 960–1460 characters  

✅ Why overlap?
👉 chunk_overlap=20

Ensures context continuity
Prevents breaking sentences abruptly
Helps retrieval quality

Characteristics of this strategy





























FeatureDescriptionBased onCharacter countContext awareness❌ NoSemantic meaning❌ IgnoredSpeed✅ FastSimplicity✅ Very simple

```python
import uuid
from pathlib import Path

from langchain_core.documents import Document
from langchain_text_splitters import RecursiveCharacterTextSplitter

from langchain_openai import OpenAIEmbeddings, ChatOpenAI
from langchain_chroma import Chroma

from langchain_core.prompts import ChatPromptTemplate
from langchain_core.output_parsers import StrOutputParser


# ---------------------------------------
# 1. Load your study material manually
#    This avoids langchain_community TextLoader warning
# ---------------------------------------

file_path = "family.txt"   # Change path if needed, e.g. "/content/family.txt"

text = Path(file_path).read_text(encoding="utf-8")

documents = [
    Document(
        page_content=text,
        metadata={"source": file_path}
    )
]

print("Original Document Loaded")
print(documents[0].page_content)


# ---------------------------------------
# 2. Define Parent and Child Splitters
# ---------------------------------------

# Parent chunks = larger chunks for full context
parent_splitter = RecursiveCharacterTextSplitter(
    chunk_size=800,
    chunk_overlap=100
)

# Child chunks = smaller chunks for accurate similarity search
child_splitter = RecursiveCharacterTextSplitter(
    chunk_size=200,
    chunk_overlap=40
)


# ---------------------------------------
# 3. Create Parent Chunks
# ---------------------------------------

parent_documents = parent_splitter.split_documents(documents)

print("\nTotal Parent Chunks:", len(parent_documents))

# This dictionary stores parent chunks using parent_id
parent_store = {}

child_documents = []


# ---------------------------------------
# 4. Create Child Chunks from each Parent
# ---------------------------------------

for parent_doc in parent_documents:
    parent_id = str(uuid.uuid4())

    # Store parent document
    parent_store[parent_id] = parent_doc

    # Split each parent into smaller child chunks
    child_chunks = child_splitter.split_documents([parent_doc])

    for child in child_chunks:
        child.metadata["parent_id"] = parent_id
        child.metadata["source"] = file_path
        child_documents.append(child)


print("Total Child Chunks:", len(child_documents))


# ---------------------------------------
# 5. Print Parent and Child Chunks
# ---------------------------------------

print("\n================ PARENT CHUNKS ================")

for i, doc in enumerate(parent_documents, start=1):
    print(f"\n--- Parent Chunk {i} ---")
    print(doc.page_content)


print("\n================ CHILD CHUNKS ================")

for i, doc in enumerate(child_documents, start=1):
    print(f"\n--- Child Chunk {i} ---")
    print("Parent ID:", doc.metadata["parent_id"])
    print(doc.page_content)


# ---------------------------------------
# 6. Create Embeddings
# ---------------------------------------

embeddings = OpenAIEmbeddings(
    model="text-embedding-3-small"
)


# ---------------------------------------
# 7. Store Child Chunks in ChromaDB
# ---------------------------------------

# Generate a unique collection name for each run to ensure a fresh database
collection_name = f"family_hierarchical_chunking_{uuid.uuid4()}"

db = Chroma.from_documents(
    documents=child_documents,
    embedding=embeddings,
    collection_name=collection_name
)

retriever = db.as_retriever(
    search_type="similarity",
    search_kwargs={"k": 3}
)


# ---------------------------------------
# 8. Retrieve Parent Context from Child Matches
# ---------------------------------------

def retrieve_parent_context(query):
    """
    Search happens on child chunks.
    Final context comes from parent chunks.
    """

    child_results = retriever.invoke(query)

    print("\n================ RETRIEVED CHILD CHUNKS ================")

    parent_ids = set()

    for i, child in enumerate(child_results, start=1):
        print(f"\n--- Retrieved Child Chunk {i} ---")
        print(child.page_content)

        parent_id = child.metadata["parent_id"]
        parent_ids.add(parent_id)

    print("\n================ MATCHED PARENT CHUNKS ================")

    parent_contexts = []

    for parent_id in parent_ids:
        parent_doc = parent_store[parent_id]
        parent_contexts.append(parent_doc.page_content)

        print("\n--- Parent Context ---")
        print(parent_doc.page_content)

    final_context = "\n\n".join(parent_contexts)

    return final_context


# ---------------------------------------
# 9. Create LLM and Prompt
# ---------------------------------------

llm = ChatOpenAI(
    model="gpt-4o-mini",
    temperature=0
)

prompt = ChatPromptTemplate.from_template("""
You are a helpful assistant.

Answer the question using only the given context.

If the answer is not available in the context, say:
"The answer is not available in the provided document."

Context:
{context}

Question:
{question}

Answer:
""")


# ---------------------------------------
# 10. RAG Function
# ---------------------------------------

def ask_question(query):
    context = retrieve_parent_context(query)

    chain = prompt | llm | StrOutputParser()

    answer = chain.invoke({
        "context": context,
        "question": query
    })

    print("\n================ FINAL ANSWER ================")
    print("Question:", query)
    print("Answer:", answer)

    return answer


# ---------------------------------------
# 11. Ask Questions
# ---------------------------------------

queries = [
    "Suhas is having how many years of experience?",
    "What is qualification of Suhas?",
    "Whether Suhas is Post Graduated?"
]

for query in queries:
    ask_question(query)
```

✅ Why TWO parent chunks are selected?
🔹 Short Answer:
👉 Because multiple child chunks matched your query, and those child chunks belong to different parent chunks

✅ More complete answer

One parent may contain partial info
Another parent completes the meaning

✅ Real-world behavior
In enterprise documents:

Info is distributed across sections
Answer requires combining multiple contexts

```python
✅ ✅ Yes — Cost ↑ but Precision ↑ (Correct ✅)
👉 In Hierarchical Chunking:
More context → More tokens → Higher cost 💰
Better context → Better accuracy → Higher precision 🎯
```

🧠 Why cost increases?
In your pipeline:

Query matches multiple child chunks
System pulls multiple parent chunks
All parent chunks are sent to LLM as context

👉 So your prompt becomes:
LLM Input = Query + Parent Chunk 1 + Parent Chunk 2 + ...

✅ More tokens = higher cost

⚖️ Trade-off (VERY IMPORTANT for training)





















FactorEffectContext size ↑Cost ↑Context richness ↑Accuracy ↑Multiple parentsCompleteness ↑

👉 Yes, hierarchical chunking + RAG challenges
➡️ are a major reason why newer LLMs have larger context windows

Query → multiple child chunks → multiple parent chunks → large context



👉 Result:

More tokens sent to LLM
Better precision
Higher cost

🎯 2. Why larger context helps
Earlier models (~4K–8K tokens):
❌ Could NOT handle:

Multiple documents,
Large RAG context,
Long conversations,


✅ Modern models (1M+ context):
👉 Can handle:

Entire PDF ✅
Multiple parent chunks ✅
Full knowledge base ✅
Multi-step agent reasoning ✅

Multiple parents → More tokens → Bigger context needed

👉 Hence:
✅ LLMs evolved to support larger context windows
